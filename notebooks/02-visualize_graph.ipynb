```python
# In a Jupyter Notebook, each commented section can be treated as a separate cell.

# --- [CELL 1] ---
# # 1. Setup and Imports
#
# This cell adds the project's root directory to the system path so we can
# import our graph application. It also imports the necessary display
# utilities from IPython.

import os
import sys
from IPython.display import Image, display

# Add the project root to the Python path
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.insert(0, project_root)
    print(f"Added '{project_root}' to the system path.")

# Import the compiled LangGraph application from our backend code
# This is the same `app` that our FastAPI server uses.
try:
    from src.backend.core.graph import app
    print("Successfully imported the LangGraph application.")
except ImportError as e:
    print(f"Failed to import the application. Error: {e}")
    print("Please ensure all backend files are correctly in place.")


# --- [CELL 2] ---
# # 2. Generate and Display the Graph Visualization
#
# Now we can call the `.get_graph()` method on our compiled app and then
# use its `.draw_mermaid_png()` method to generate a visual representation
# of our agentic workflow.

if 'app' in locals():
    try:
        # Generate the PNG image data of the graph
        png_data = app.get_graph().draw_mermaid_png()
        
        # Display the image directly in the notebook
        display(Image(png_data))
        
        print("Graph visualization generated successfully.")
        
    except Exception as e:
        print(f"An error occurred while generating the graph visualization: {e}")
        print("Please ensure pygraphviz is installed and accessible in your system's PATH if you are on Windows.")
else:
    print("Application object 'app' not found. Cannot generate graph.")

```
